# here combined list of genres

```python
def fun(x):
  L=[]
  for i in range(len(x)):
        L.append(x[i]["name"])
  return L
for j in range(0,df["genres"].count()):
    df["genres"][j]=fun(df["genres"][j])
```

# here we removed spaces from strings in list

```python
def fun(x):
    for i in range(len(x)):
        L=[]
        L=x[i].split(" ")
        x[i]="".join(L)

for j in range(0,df["genres"].count()):
   fun(df["genres"][j])
```
